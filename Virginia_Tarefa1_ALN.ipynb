## Círculo Unitário

<p> A equação do circulo é dada como $ r = x^{2} + y^{2} $ .</p>

<p> Utilizando a parametrização é possivel obter x e y em função do raio e do ângulo. </p>

<p> $ x = r \cdot cos(\theta) $ </p>

<p> $ y = r \cdot sin(\theta) $ </p>

<p> Em forma vetorial temos: </p>

<p> $ C = \begin{bmatrix} r \cdot cos(\theta) \\ r \cdot sin(\theta) \end{bmatrix} $ </p>

<p> O círculo unitário possui raio igual a 1. Portanto pode ser escrito da forma <p/>

<p> $ C = \begin{bmatrix} cos(\theta) \\ sin(\theta) \end{bmatrix} $ </p>

##### plotando o círculo unitário C

```python
theta = np.arange(0, 2*m.pi + 0.001, 0.001)
C = np.array([np.cos(theta),  np.sin(theta)])
r = 1

plt.axis('equal')
plt.plot(C[0], C[1])
plt.show()
```

## Aplicando uma Transformação

<p> Seja a matriz transformação $ T = \begin{bmatrix} 2 & 3 \\ 4 & 1 \end{bmatrix} $ é possível aplicar a transformação T em C fazendo $ C_{tranform} = (C^{T} \cdot  T)^{T} = T \cdot C $ .</p>



##### aplicando a transformação T no círculo C

```python
T = np.array([[2, 3],  [4, 1]])

CxT = np.matmul(T,C)

plt.axis('equal')
plt.plot(CxT[0], CxT[1])
plt.show()
```

## Decomposição em Valores Singulares (SVD)

<p> Decomponto T em valores singulares obtemos as matrizes de rotação $ U $ e $ V^{T} $, e a matriz de esticamento $ Σ $ .  </p>

<p> $ U = \begin{bmatrix}-0.64074744 & -0.76775173 \\ -0.76775173 & 0.64074744 \end{bmatrix} $ </p>
<p> $ Σ = \begin{bmatrix} 5.11667274 & 0 \\ 0 & 1.95439508 \end{bmatrix} $ </p>
<p> $ V^{T} = \begin{bmatrix}-0.85065081 & -0.52573111 \\ 0.52573111 & -0.85065081 \end{bmatrix} $ </p>

##### calculando SVD

```python
U, s, V = np.linalg.svd(T, full_matrices=True)
E = np.diag(s)

print("U = ", U, "\n")
print("Σ = ", E, "\n")
print("V* = ", V, "\n")
```

### Aplicando a rotação $ V^{T} $

```python
CxV = np.matmul(V,C)

plt.axis('equal')
plt.plot(CxV[0], CxV[1])
plt.show()
```

### Aplicando o esticamento $ Σ $

```python
CxVxE = np.matmul(E,CxV)

plt.axis('equal')
plt.plot(CxVxE[0], CxVxE[1])
plt.show()
```

### Aplicando a rotação $ U $

```python
CxVxExU = np.matmul(U,CxVxE)

plt.axis('equal')
plt.plot(CxVxExU[0], CxVxExU[1])
plt.show()
```

## Aplicações Práticas

<p> Com base nos resultdos deste trabalho é possível imaginar que a modelagem de matrizes pode ter diversas aplicações na área de processamento de imagens. Transformações como as aplicadas aqui no círculo unitário podem ser também aplicadas em imagens, que nada mais são que matrizes. </p>

<p> Como exemplo podemos aplicar a matriz de rotação de 90° a uma imagem escolhida no pixabay. (Fonte: https://pixabay.com/en/staniel-cay-swimming-pig-seagull-171908/) </p>

<p> $ T_{rotação90°} = \begin{bmatrix} cos(\frac{\pi}{2}) & sin(\frac{\pi}{2}) \\ -sin(\frac{\pi}{2}) & cos(\frac{\pi}{2}) \end{bmatrix} $ </p>

```python
# Vamos rotacionar o porco.

# Ler a imagem a ser rotacionada.
PIG = img.imread("pig.jpg")

# Atribuir a T6 a matriz de rotação de 90°.
T6 = np.array([[m.cos(m.pi/2), m.sin(m.pi/2)],  [-m.sin(m.pi/2), m.cos(m.pi/2)]])

# Atribuir a T6_INV a inversa da transformação de rotação de 90°.
T6_INV = np.linalg.inv(T6)

# Aplicar a transformação nas dimensões da imagem para
# saber quais as dimensões da nova imagem.
size = np.matmul(T6,[[len(PIG[0])],[len(PIG)]])
h = abs(int(size[1][0]))
w = abs(int(size[0][0]))

# Criar a nova imagem só com pixels pretos.
PIGxT6 = np.zeros((h,w,3),dtype=np.uint8)

# Percorrer todos os pixels da nova imagem.
for i in range(0,h):
    for j in range(0,w):
        
        # Aplicar a transformação inversa no ponto da nova imagem para 
        # descobrir a qual ponto da imagem antiga ele corresponde.
        pixel = np.matmul(T6_INV,[[j],[i]])
        
        # Atribuir valores R, G e B da nova imagem.
        PIGxT6[i,j,0] = PIG[int(pixel[1][0]),int(pixel[0][0]),0] # R
        PIGxT6[i,j,1] = PIG[int(pixel[1][0]),int(pixel[0][0]),1] # G
        PIGxT6[i,j,2] = PIG[int(pixel[1][0]),int(pixel[0][0]),2] # B

# Plotar a imagem original e a rotacionada.
f, (plt1, plt2) = plt.subplots(1, 2, figsize=(10,5))

plt1.set_title('Original')
plt1.axis('equal')
plt1.imshow(PIG)        

plt2.set_title('Rotacionado')
plt2.axis('equal')        
plt2.imshow(PIGxT6)

plt.show()

# O porco foi rotacionado com sucesso! :3
```
