## Coordenadas Homogêneas

<p> Um ponto $ \left(x,y\right) $ no plano cartesiano é representado em coordenadas homogêneas como $ \left(wx, wy, w\right) $. Em outras palavras, seja a coordenada homogênea $ \left(x', y', w\right) $ temos que o ponto no plano cartesiano o qual ela representa é $ \left(x, y\right) $ onde $ x = \frac{x'}{w} $ e $ y = \frac{x'}{w} $. Podemos admitir $ w = 1 $ e dessa forma escrevermos $ \left(x, y\right) $ como $ \left(x, y, 1\right) $.</p>

## Círculo unitário em coordenadas homogêneas

<p> Da tarefa 1 temos que o círculo unitário, $ C $, pode ser representado como 

$ \begin{bmatrix} cos\left(\theta\right) \\ sin\left(\theta\right) \end{bmatrix} $

. Em coordenadas homogêneas temos, portanto, temos: </p>

$ C = \begin{bmatrix} cos\left(\theta\right) \\ sin\left(\theta\right) \\ 1 \end{bmatrix} $

## Matriz de Transformação 3x3

<p> Dada matriz de transformação </p>

$ T = \begin{bmatrix} a & b & A \\ c & d & B \\ 0 & 0 & 1 \end{bmatrix} $

<p> $a$, $b$, $c$ e $d$ correspondem a matriz de transformação 2x2, que conferem uma transformação linear (rotação e esticamento) a um vetor. Os valores $A$ e $B$ são os que transladam um vetor em x e y respetivamente. </p>

<p> Usando  o círculo unitário representado em coordenadas homogeneas e aplicando a transformação $T$ em $C$ temos: </p>

$ T \cdot C =  $
$ \begin{bmatrix} a & b & A \\ c & d & B \\ 0 & 0 & 1 \end{bmatrix} $ 
$ \cdot $
$ \begin{bmatrix} cos(\theta) \\ sin(\theta) \\ 1 \end{bmatrix} $
$ = $
$ \begin{bmatrix} a \cdot cos(\theta) + b \cdot sin(\theta) + A \\ c \cdot cos(\theta) + d \cdot sin(\theta) + B \\ 1 \end{bmatrix} $

<p> Na tarefa 1 a rotação e o esticamente foi bem trabalhado, portanto vamos observar uma nova transformação onde apenas a translação é aplicada. </p>

<p> Uma matriz de translação é dada por </p>

$ T_{translação} = \begin{bmatrix} 1 & 0 & A \\ 0 & 1 & B \\ 0 & 0 & 1 \end{bmatrix} $

<p> Ao aplicarmos a translação no círculo unitário temos: </p>

$ T_{translação} \cdot C = $
$ \begin{bmatrix} 1 & 0 & A \\ 0 & 1 & B \\ 0 & 0 & 1 \end{bmatrix} $
$ \cdot $
$ \begin{bmatrix} cos(\theta) \\ sin(\theta) \\ 1 \end{bmatrix} $
$ = $
$ \begin{bmatrix} cos(\theta) + A \\ sin(\theta) + B \\ 1 \end{bmatrix} $

## Aplicando uma Translação

<p> Seja a matriz de transformação $ T = \begin{bmatrix} 1 & 0 & 3 \\ 0 & 1 & 2 \\ 0 & 0 & 1 \end{bmatrix} $, vamos aplicá-la ao círculo unitário $ C = \begin{bmatrix} cos(\theta) \\ sin(\theta) \\ 1 \end{bmatrix} $. Dessa forma obtemos $ C_{transform} = \begin{bmatrix} cos(\theta) + 3 \\ sin(\theta) + 2 \\ 1 \end{bmatrix} $. </p>

##### plotanto $C$ e $C_{transform}$

```python
# Círculo unitário
theta = np.arange(0, 2*m.pi + 0.001, 0.001)
C = np.array([np.cos(theta),  np.sin(theta), np.ones(len(theta))])

# Transformação
T = np.array([[1, 0, 3],  [0, 1, 2], [0, 0, 1]])

# T.C
TxC = np.matmul(T,C)

# PLOTANDO:

# Tamanho e limites do gráfico
plt.figure(figsize=(5,5))
plt.xlim(xmin=-2, xmax=5)
plt.ylim(ymin=-2, ymax=5)

# Plotando C
plt.plot(C[0]/C[2], C[1]/C[2], "--b")

# Plotando o centro de C
plt.plot(0,0,"ob")
plt.plot([0,-2],[0,0],"--b")
plt.plot([0,0],[0,-2],"--b")

# Plotando C_transform
plt.plot(TxC[0]/TxC[2], TxC[1]/TxC[2], "r")

# Plotando o centro de C_transform
plt.plot(3,2,"or")
plt.plot([3,-2],[2,2],"--r")
plt.plot([3,3],[2,-2],"--r")

# Plotando a linha da translação
plt.plot([0,3],[0,2],"--k")

# Legenda
blue_patch = mpatches.Patch(color='blue', label='C')
red_patch = mpatches.Patch(color='red', label='C_transform')
black_patch = mpatches.Patch(color='black', label='Translation line')
plt.legend(handles=[blue_patch, red_patch, black_patch])

# Mostrar o gráfico
plt.show()
```

## Decomposição em Valores Singulares (SVD) de uma matriz de transformação linear

<p> Seja $ T = \begin{bmatrix} 2cos\left(60^{o}\right) & -3sin\left(60^{o}\right) & 0 \\ 2sin\left(60^{o}\right) & 3cos\left(60^{o}\right) & 0 \\ 0 & 0 & 1 \end{bmatrix} $ , T é uma transformação linear que aplica uma rotação de $ 60^{o} $, um esticamento de $ 2 $ no eixo $ x $ e um esticamento de $ 3 $ no eixo $ y $. </p>

<p> Decompondo T em valores singulares obtemos </p>

<p> $ U = \begin{bmatrix} cos\left(150^{o}\right) & -sin\left(150^{o}\right) & 0 \\ sin\left(150^{o}\right) & cos\left(150^{o}\right) & 0 \\ 0 & 0 & 1 \end{bmatrix} $ </p>
<p> $ Σ = \begin{bmatrix} 3 & 0 & 0 \\ 0 & 2 & 0 \\ 0 & 0 & 1 \end{bmatrix} $ </p>
<p> $ V^{T} = \begin{bmatrix} cos\left(270^{o}\right) & -sin\left(270^{o}\right) & 0 \\ sin\left(270^{o}\right) & cos\left(270^{o}\right) & 0 \\ 0 & 0 & 1 \end{bmatrix} $ </p>

##### calculando SVD e plotando

```python
# Círculo unitário
theta = np.arange(0, 2*m.pi + 0.001, 0.001)
C = np.array([np.cos(theta),  np.sin(theta), np.ones(len(theta))])

# Transformação
T = np.array([[2*m.cos(m.pi/3), -3*m.sin(m.pi/3), 0],  [2*m.sin(m.pi/3), 3*m.cos(m.pi/3), 0], [0, 0, 1]])

# SVD
U, s, V = np.linalg.svd(T, full_matrices=True)
E = np.diag(s)

print("U = ", U, "\n")
print("Σ = ", E, "\n")
print("V* = ", V, "\n")

# Multiplicações
VxC = np.matmul(V,C)
ExVxC = np.matmul(E,VxC)
UxExVxC = np.matmul(U,ExVxC)

# Plots
f, (plt0, plt1, plt2, plt3) = plt.subplots(1, 4, figsize=(20,5))

plt0.set_title('C')
plt0.axis('equal')
plt0.plot(C[0]/C[2], C[1]/C[2])
plt0.plot(C[0][0]/C[2][0], C[1][0]/C[2][0], "o")

plt1.set_title('VxC')
plt1.axis('equal')
plt1.plot(VxC[0]/VxC[2], VxC[1]/VxC[2])
plt1.plot(VxC[0][0]/VxC[2][0], VxC[1][0]/VxC[2][0], "o")

plt2.set_title('ExVxC')
plt2.axis('equal')
plt2.plot(ExVxC[0]/ExVxC[2], ExVxC[1]/ExVxC[2])
plt2.plot(ExVxC[0][0]/ExVxC[2][0], ExVxC[1][0]/ExVxC[2][0], "o")

plt3.set_title('UxExVxC')
plt3.axis('equal')
plt3.plot(UxExVxC[0]/UxExVxC[2], UxExVxC[1]/UxExVxC[2])
plt3.plot(UxExVxC[0][0]/UxExVxC[2][0], UxExVxC[1][0]/UxExVxC[2][0], "o")

plt.show()
```

## Decomposição em Valores Singulares (SVD) de uma matriz de transformação não-linear (com translação)

<p> Seja $ T = \begin{bmatrix} 2cos\left(60^{o}\right) & -3sin\left(60^{o}\right) & 2 \\ 2sin\left(60^{o}\right) & 3cos\left(60^{o}\right) & 3 \\ 0 & 0 & 1 \end{bmatrix} $ , T é uma transformação que aplica uma rotação de $ 60^{o} $, um esticamento de $ 2 $ no eixo $ x $, um esticamento de $ 3 $ no eixo $ y $, uma translação de $ 2 $ unidades no eixo $ x $ e uma translação de $ 3 $ unidades no eixo $ y $. Por conta da translação, $ T $ não é uma transformação linear. </p>

<p> Ao tentar deconpor T em valores singulares obtemos </p>

<p> $ U = \begin{bmatrix} -0.57159536 & -0.81311649 & 0.11009233 \\ -0.79330453 & 0.58190783 & 0.17902847 \\ -0.20963459 & 0.0149951 & -0.97766481 \end{bmatrix} $ </p>
<p> $ Σ = \begin{bmatrix} 4.21970738 & 0 & 0 \\ 0 & 2.99476875 & 0 \\ 0 & 0 & 0.47479444 \end{bmatrix} $ </p>
<p> $ V^{T} = \begin{bmatrix} -0.4610839 & 0.06993175 & -0.88459663 \\ 0.06503923 & 0.99687174 & 0.04490684 \\ 0.8849698 & -0.03682766 & -0.46418981 \end{bmatrix} $ </p>

```python
# Círculo unitário
theta = np.arange(0, 2*m.pi + 0.001, 0.001)
C = np.array([np.cos(theta),  np.sin(theta), np.ones(len(theta))])

# Transformação
T = np.array([[2*m.cos(m.pi/3), -3*m.sin(m.pi/3), 2],  [2*m.sin(m.pi/3), 3*m.cos(m.pi/3), 3], [0, 0, 1]])

# SVD
U, s, V = np.linalg.svd(T, full_matrices=True)
E = np.diag(s)

print("U = ", U, "\n")
print("Σ = ", E, "\n")
print("V* = ", V, "\n")

# Multiplicações
VxC = np.matmul(V,C)
ExVxC = np.matmul(E,VxC)
UxExVxC = np.matmul(U,ExVxC)

# Plots
f, (plt0, plt1, plt2, plt3) = plt.subplots(1, 4, figsize=(20,5))

plt0.set_title('C')
plt0.axis('equal')
plt0.plot(C[0]/C[2], C[1]/C[2])
plt0.plot(C[0][0]/C[2][0], C[1][0]/C[2][0], "o")

plt1.set_title('VxC')
plt1.axis('equal')
plt1.plot(VxC[0]/VxC[2], VxC[1]/VxC[2])
plt1.plot(VxC[0][0]/VxC[2][0], VxC[1][0]/VxC[2][0], "o")

plt2.set_title('ExVxC')
plt2.axis('equal')
plt2.plot(ExVxC[0]/ExVxC[2], ExVxC[1]/ExVxC[2])
plt2.plot(ExVxC[0][0]/ExVxC[2][0], ExVxC[1][0]/ExVxC[2][0], "o")

plt3.set_title('UxExVxC')
plt3.axis('equal')
plt3.plot(UxExVxC[0]/UxExVxC[2], UxExVxC[1]/UxExVxC[2])
plt3.plot(UxExVxC[0][0]/UxExVxC[2][0], UxExVxC[1][0]/UxExVxC[2][0], "o")

plt.show()

print("\n\n VxC e ExVxC discreto")

g, (plt4, plt5) = plt.subplots(1, 2, figsize=(10,5))

plt4.set_title('VxC')
plt4.axis('equal')
plt4.plot(VxC[0]/VxC[2], VxC[1]/VxC[2], ".")
plt4.plot(VxC[0][0]/VxC[2][0], VxC[1][0]/VxC[2][0], "o")

plt5.set_title('ExVxC')
plt5.axis('equal')
plt5.plot(ExVxC[0]/ExVxC[2], ExVxC[1]/ExVxC[2], ".")
plt5.plot(ExVxC[0][0]/ExVxC[2][0], ExVxC[1][0]/ExVxC[2][0], "o")

plt.show()
```

<p> Por não se tratar de uma transformação linear, o SVD falhou para $ T $ quanto aos tipos de matrizes esperados para $ u $, $ \Sigma $ e $ v^{T} $. É possível perceber isso com maior facilidade ao observar os gráficos, quando o círculo se transforma em duas retas ao ser multiplicado por $ v^{T} $. Ao decompor uma transformação utilizando o SVD obtemos duas matrizes de rotação e uma de esticamento<b>*</b>, que são transformações lineares, uma translação, que não é transformação linear, não é esperada. </p>

<p> <b>*</b> Esticamento, contração ou espelhamento. </p>

<p> Uma forma de solucionar o problema é separar a translação das outras transformações (esticamento e rotações) e aplicar o SVD apenas na transformação linear. Teriamos, portanto, </p>

<p> $$ T = T_{translação} \cdot T_{linear} = T_{translação} \cdot u \cdot \Sigma \cdot v^{T} . $$ </p>
