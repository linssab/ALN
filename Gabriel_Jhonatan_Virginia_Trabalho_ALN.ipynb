```python
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.patches as mpatches
import random
import matplotlib.image as mpimg
from scipy.ndimage import convolve
```

```python
INF = 9999999999999
PATH = 'recdev/easy/'
TEST_SET = {'all': []}
TRAINING_SET = {'2': [], '11': [], '15': [], '16': [], '87': []}

def read_image(file_name):
    return mpimg.imread(PATH + file_name)

def load(): 
    for c in TRAINING_SET:
        a = random.randrange(8)
        b = random.randrange(8)     
        while b == a:
            b = random.randrange(8)
        for i in range(0,8):
            file_name = c + '-' + str(i+1) + '.jpg'
            if i == a or i == b:
                TEST_SET['all'] = np.ndarray.tolist(np.append(file_name, TEST_SET['all']))
            else:
                TRAINING_SET[c] = np.ndarray.tolist(np.append(file_name, TRAINING_SET[c]))

def compare(img1,img2):
    return np.linalg.norm(img1-img2)

def find_class(img):
    mn = INF
    mi = []
    mc = 0
    for c in TRAINING_SET:
        for i in range(0,6):
            img_compare = read_image(TRAINING_SET[c][i])
            mn_candidate = compare(img,img_compare)
            if mn_candidate < mn:
                mn = mn_candidate
                mi = img_compare
                mc = c
    return mn, mi, mc

def main():
    load()

    for fn in TEST_SET['all']:
        img = read_image(fn)
        mn, mi, mc = find_class(img)
        print('min_norm_difference = ' + str(mn))
        print('class = ' + str(mc))
        print('mi =')
        f, (plt0, plt1) = plt.subplots(1, 2, figsize=(10,5))
        plt0.imshow(img)
        plt1.imshow(mi)
        plt.show()

main()
```
